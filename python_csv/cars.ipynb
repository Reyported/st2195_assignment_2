```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.71 Safari/537.36'}
cars = requests.get("https://en.wikipedia.org/wiki/Comma-separated_values", headers=headers)
```

```python
stonks = BeautifulSoup(cars.content, 'html.parser')
```

```python
print(stonks.prettify())
```

```python
cartable = stonks.find_all('table', class_='wikitable')
```

```python
cartable
```

```python
cardf = pd.read_html(str(cartable))

cardf=pd.DataFrame(cardf[0])
print(cardf.head())
```

```python
cardf.to_csv('cars.csv')
```
